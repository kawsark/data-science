## Initialize Notebook and load data

```python
# implement linear regression using the pandas and matplotlib libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the data.csv file as a pandas dataframe
data = pd.read_csv("data.csv")

# print the dataframe
print(data)

# get information about the dataframe
```

## Cleanup and visualize data

```python
# Drop the rows with null values
data = data.dropna()

# Get x values from the sq. ft. column
x = data["sq. ft."]

# Get y values from the Price column
y = data["Price"]

# plot the x and y values
plt.plot(x, y, "o")

# Label the x and y axes
plt.xlabel("Sq. Ft.")
```

## Create model and fit to data

```python
# Create a liner regression model using the sklearn library
from sklearn.linear_model import LinearRegression

# Define a function to fit the linear regression model
def fit_linear_regression(x, y):
    # Create a linear regression model
    model = LinearRegression()
    # Fit the model
    model.fit(x.values.reshape(-1, 1), y)
    # Return the model
    return model

# call the fit_linear_regression function and pass the x and y values
model = fit_linear_regression(x, y)
```

## Plot original data and the regression line

```python
# Plot the original x and y values 
plt.plot(x, y, "o")

# Plot the regression line
plt.plot(x, model.predict(x.values.reshape(-1, 1)))

# Label the axes
plt.xlabel("Sq. Ft.")
plt.ylabel("Price")
plt.show()
```

## Predict price using regression line

```python
# get the line's slope and intercept
slope = model.coef_[0]

# Get the predicted home price from the model for a house with 5000 sq. ft.
predicted_price = model.predict([[5000]])[0]

# print the slope and predicted price
print(slope, predicted_price)

# write 5000 sq. ft. and predicted price list to a result.csv file
result = pd.DataFrame({"sq. ft.": [5000], "predicted price": [predicted_price]})
result.to_csv("result.csv", index=False)
```
